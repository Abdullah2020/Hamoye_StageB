**Question 12:**

From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
#let assign x to T2 and y to T6

x = energy_df["T2"]
y = energy_df["T6"]
```

```python
#in order to fit our linear model, we have to reshape our x variable

arr = np.array(x)
x_feature = arr.reshape(-1, 1)
```

```python
#we can now split our data into train and test set

x_train, x_test, y_train, y_test = train_test_split(x_feature, y, train_size=0.7, test_size = 0.3, random_state=42)

#Define our model
linear_m = LinearRegression()

#fit the model
linear_m.fit(x_train, y_train)

#predict the model
preds = linear_m.predict(x_test)
```

```python
#finally, we can now compute for R_squared value

r2 = r2_score(y_test, preds)
print('The value of R_squared is :', round(r2, 2))
```

**Question 13:**

Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?

```python
#let remove date and lights columns

new_energy_df = energy_df.drop(['lights', 'date'], axis = 1)
```

```python
new_energy_df.head()
```

```python
#we normalise our dataset to a common scale using the Min Max Scaler

scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(new_energy_df), columns = new_energy_df.columns)

features_df = normalised_df.drop(columns = ['Appliances'], axis=1)
energy_target = normalised_df['Appliances']
```

```python
#Use a 70-30 train-test set split with a random state of 42 (for reproducibility)

x_train, x_test, y_train, y_test = train_test_split(features_df, energy_target, train_size=0.7, test_size = 0.3, random_state=42)

#Define the model
linear_model = LinearRegression()

#fit our model
linear_model.fit(x_train, y_train)

#prediction
predicted_values = linear_model.predict(x_test)
```

```python
#The Mean Absolute Error(MAE) in two decimal places

df_mae = mean_absolute_error(y_test, predicted_values)

print("The Mean Absolute Error (MAE) is: ", round(df_mae,2))
```

**Question 14:**

What is the Residual Sum of Squares (in two decimal places)?

```python
residual_ss = np.sum(np.square(y_test - predicted_values))

print("The Residual Sum Squares (RSS) is: ", round(residual_ss, 2))
```

**Question 15:**

What is the Root Mean Squared Error (in three decimal places)?

```python

result_ms_err = np.sqrt(mean_squared_error(y_test, predicted_values))

print("The Root Mean Squared Error is : ", round(result_ms_err, 3))
```

**Question 16:**

What is the Coefficient of Determination (in two decimal places)?

```python
R_square = r2_score(y_test, predicted_values)

print("The Coefficient of Determination is ", round(R_square,2))
```

**Question 17:**

Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_weights_df(model, feat, col_name):

  weights = pd.Series(model.coef_, feat.columns).sort_values()

  weights_df = pd.DataFrame(weights).reset_index()

  weights_df.columns = ['Features', col_name]
  
  weights_df[col_name].round(3)
  return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```

```python
linear_model_weights
```

```python
#the feature with the highest weight

print(linear_model_weights[linear_model_weights.Linear_Model_Weight == linear_model_weights.Linear_Model_Weight.max()])
```

```python
#the feature with the lowest weight

print(linear_model_weights[linear_model_weights.Linear_Model_Weight == linear_model_weights.Linear_Model_Weight.min()])
```

**Question 18:**

Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
#Ridge Regression

#train ridge regression model where alpha is 0.4
l2_reg = Ridge(alpha=0.4, random_state=None)

#fit the model
l2_reg.fit(x_train, y_train)

#predict the model
l2_preds = l2_reg.predict(x_test)
```

```python
#get the Root Mean Squared Error (RMSE)

l2_rmse = np.sqrt(mean_squared_error(y_test, l2_preds))

print("The Root Mean Squared Error (RMSE) is: ", round(l2_rmse, 3))
```

The Root Mean Squared Error (RMSE) on Ridge Regression model and Linear model is the same and equal to 0.088. Thus, there is no changes in the RMSE

Question 19:

Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
#Lesson Regression model with alpha = 0.001

#Define the model
lasso_reg = Lasso(alpha=0.001)

#fit the model
lasso_reg.fit(x_train, y_train)

#predict the model
l1_pred = lasso_reg.predict(x_test)
```

```python
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
```

```python
lasso_weights_df
```

```python
#How many of the features have non-zero feature weights?

non_z = lasso_weights_df["Lasso_weight"].astype(bool).sum(axis=0)

print("The total features that have non-zero feature weights is = ", non_z)
```

**Question 20:**

What is the new RMSE with the lasso regression? (Answer should be in three (3) decimal places)

```python
#The new Root Mean Squared Error (RMSE) with LASSO

new_lasso_rmse = np.sqrt(mean_squared_error(y_test, l1_pred))

print("The New Root Mean Squared Error (RMSE) on LASSO is: ", round(new_lasso_rmse, 3))
```
